```python
import requests
from bs4 import BeautifulSoup
import csv
import time
```

```python
# Ulta CSV

csvfile = open('ulta_products.csv', 'w', newline='', encoding='utf-8')

c = csv.writer(csvfile)

# c.writerow( ['name', 'brand', 'low', 'high'] )

url = 'https://www.ulta.com/shop/skin-care/all?gn=PREFERENCE--vegan&page='
for x in range(5):
    page = requests.get(url + str(x))
    soup = BeautifulSoup(page.text, 'html.parser')

    product_info = soup.find_all('li')

    # Loop to get stuff from each product one the page
    for x in product_info:
        try:
            name = x.find('span', class_='ProductCard__product').get_text()
            brand = x.find('span', class_='ProductCard__brand').get_text()[0:-2]
            price = x.find('div', class_='ProductCard__price').get_text()
            kit = price.find('Kit')
            if kit > -1:
                kit = kit.split('Kit')[0]
            sale = price.find('sale')
            if sale > -1:
                sale = sale.split('sale')[0]
            try:
                price = price.split('-')
                low = price[0].strip().replace('$', '')
                high = price[1].strip().replace('$', '')
                c.writerow( [name, brand, low, high] )
            except:
                low = price
                high = price
                low = low[0].replace('$', '')
                high = high[0].replace('$', '')
                c.writerow( [name, brand, low, high] )
            # print( [product_name, product_brand, product_low, product_high] )
        except:
            pass

csvfile.close()
```

```python
# r.e.m. beauty CSV

csvfile = open('rem_products.csv', 'w', newline='', encoding='utf-8')

c = csv.writer(csvfile)

for x in range(1):
    page = requests.get("https://rembeauty.com/collections/all-treatments")
    soup = BeautifulSoup(page.text, 'html.parser')
    
    product_info = soup.find_all('div', class_='tvg_product_card_rTJCUa_ProductCard-Content js-card-content')
    
    # Loop to get stuff from each product one the page
    for x in product_info:
        try:
            name = x.find('p', class_='tvg_product_card_rTJCUa_ProductCard-ContentTitle').get_text()
            brand = 'r.e.m. beauty'
            price = x.find('p', class_='tvg_product_card_rTJCUa_ProductCard-ContentPrice').get_text()
            try:
                low = price.replace('$', '')
                high = price.replace('$', '')
                c.writerow( [name, brand, low, high] )
            except:
                pass
                c.writerow( [name, brand, low, high] )
            # print( [name, brand, low, high] )
        except:
            pass

csvfile.close()
```

```python
# Pacifica

csvfile = open('pacifica_products.csv', 'w', newline='', encoding='utf-8')
c = csv.writer(csvfile)

url = 'https://www.pacificabeauty.com/collections/skincare'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

product_info = soup.find_all('product-item', class_='product-item')

for x in product_info:
    try:
        name = x.find('div', class_='product-item-meta__title').get_text().strip()
        brand = 'Pacifica'
        price_tag = x.find('span', class_='price')
        for x in price_tag.find_all(class_='visually-hidden'):
            x.extract()
        price = price_tag.get_text().strip()
        if 'Starting at' in price:
            price = price.replace('Starting at ').strip()
        else:
            pass
        if '-' in price:
            low = price.replace('$', '').strip()
            high = price.replace('$', '').strip()
        else:
            low = price.replace('$', '').strip()
            high = price.replace('$', '').strip()
        c.writerow([name, brand, low, high])
    except:
        pass

csvfile.close()
```
